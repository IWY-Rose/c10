We will import the following libraries for the lab


```python
# Pandas is a software library written for the Python programming language for data manipulation and analysis.
import pandas as pd
# NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays
import numpy as np
# Matplotlib is a plotting library for python and pyplot gives us a MatLab like plotting framework. We will use this in our plotter function to plot data.
import matplotlib.pyplot as plt
#Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics
import seaborn as sns
# Preprocessing allows us to standarsize our data
from sklearn import preprocessing
# Allows us to split our data into training and testing data
from sklearn.model_selection import train_test_split
# Allows us to test parameters of classification algorithms and find the best one
from sklearn.model_selection import GridSearchCV
# Logistic Regression classification algorithm
from sklearn.linear_model import LogisticRegression
# Support Vector Machine classification algorithm
from sklearn.svm import SVC
# Decision Tree classification algorithm
from sklearn.tree import DecisionTreeClassifier
# K Nearest Neighbors classification algorithm
from sklearn.neighbors import KNeighborsClassifier
```

This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

## TASK  2


Standardize the data in <code>X</code> then reassign it to the variable  <code>X</code> using the transform provided below.


```python
transform = preprocessing.StandardScaler()
X = transform.fit_transform(X)
```

## TASK  3


Use the function train_test_split to split the data X and Y into training and test data. Set the parameter test_size to  0.2 and random_state to 2. The training data and test data should be assigned to the following labels.

## TASK  12


Find the method performs best:


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Scores from cross-validation
cv_scores = {
    'Logistic Regression': logreg_cv.best_score_,
    'SVM': svm_cv.best_score_,
    'Decision Tree': tree_cv.best_score_,
    'KNN': knn_cv.best_score_
}

# Scores on the test set
test_scores = {
    'Logistic Regression': logreg_cv.score(X_test, Y_test),
    'SVM': svm_cv.score(X_test, Y_test),
    'Decision Tree': tree_cv.score(X_test, Y_test),
    'KNN': knn_cv.score(X_test, Y_test)
}

# Create a DataFrame for a summary report
report = pd.DataFrame({
    'Algorithm': list(cv_scores.keys()),
    'CV Best Score': list(cv_scores.values()),
    'Test Score': list(test_scores.values())
})

print(report)

best_model = report.loc[report['Test Score'].idxmax()]
print("\nBest Performing Model:")
print(best_model)

# Create grouped bar chart
plt.figure(figsize=(12, 7))

algorithms = list(cv_scores.keys())
cv_scores_values = list(cv_scores.values())
test_scores_values = list(test_scores.values())

# Set the width of the bars and positions
bar_width = 0.35
x = np.arange(len(algorithms))

# Create the bars
bars1 = plt.bar(x - bar_width/2, cv_scores_values, bar_width, 
                label='CV Best Score', color='skyblue', alpha=0.8)
bars2 = plt.bar(x + bar_width/2, test_scores_values, bar_width,
                label='Test Score', color='lightcoral', alpha=0.8)

# Customize the plot
plt.title('Classification Model Accuracy Comparison\n(CV Best Score vs Test Score)', 
          fontsize=16, fontweight='bold')
plt.xlabel('Classification Models', fontsize=12)
plt.ylabel('Accuracy Score', fontsize=12)
plt.ylim(0, 1)
plt.xticks(x, algorithms, rotation=45)

# Add value labels on top of bars
for bar, score in zip(bars1, cv_scores_values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{score:.3f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

for bar, score in zip(bars2, test_scores_values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{score:.3f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

# Add legend and grid
plt.legend(loc='upper right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate predictions using the best Logistic Regression model
yhat = logreg_cv.predict(X_test)

# Create confusion matrix
cm = confusion_matrix(Y_test, yhat)

# Enhanced visualization
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Did not land', 'Landed'], 
            yticklabels=['Did not land', 'Landed'],
            cbar_kws={'label': 'Count'})

plt.title('Confusion Matrix - Logistic Regression\n(Best Performing Model - Test Accuracy: 83.3%)', 
          fontsize=14, fontweight='bold')
plt.xlabel('Predicted Labels', fontsize=12)
plt.ylabel('True Labels', fontsize=12)

# Extract values for detailed analysis
tn, fp, fn, tp = cm.ravel()

# Add text annotations below the plot
plt.figtext(0.5, 0.02, 
           f'True Negatives (TN): {tn}  |  False Positives (FP): {fp}  |  False Negatives (FN): {fn}  |  True Positives (TP): {tp}', 
           ha='center', fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgray"))

plt.tight_layout()
plt.show()

# Print the confusion matrix values
print("Confusion Matrix Breakdown:")
print(f"True Negatives (TN): {tn} - Correctly predicted failed landings")
print(f"False Positives (FP): {fp} - Incorrectly predicted successful landings (Type I Error)")
print(f"False Negatives (FN): {fn} - Incorrectly predicted failed landings (Type II Error)")
print(f"True Positives (TP): {tp} - Correctly predicted successful landings")
print(f"\nTest Accuracy: {(tp + tn) / (tp + tn + fp + fn):.3f}")
```
